```python
import pandas as pd
```

```python
df=pd.read_csv('../data/cleaningv1.csv')
```

```python
usage_cols = ['change_mou', 'change_rev']

for col in usage_cols:
    if col in df.columns:
        # ملء القيم المفقودة بالوسيط لحماية العمودين من الانحياز بسبب القيم الشاذة
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
```

```python
df[['change_mou', 'change_rev','churn']].corr()
```

```python
garbage_cols = [
        'ethnic', 'infobase', 'HHstatin', 'dwllsize', 'dwlltype',
        'kid0_2', 'kid3_5', 'kid6_10', 'kid11_15', 'kid16_17'
    ]
df = df.drop(columns=[col for col in garbage_cols if col in df.columns])
```
